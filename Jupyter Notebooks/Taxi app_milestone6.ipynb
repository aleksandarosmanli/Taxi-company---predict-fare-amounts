##### **Target**

It can be noticed that there isn't a column that indicates tip percent, which is what I need to create the target variable. I'll have to engineer it. 

Let's add a `tip_percent` column to the dataframe by performing the following calculation:  

tip_percent = tip_amount \ (total_amount - tip_amount)  

I'll round the result to three places beyond the decimal. This is an important step, because it affects how many customers are labeled as generous tippers. In fact, without performing this step, approximately 1,800 people who do tip â‰¥ 20% would be labeled as not generous. 

```python
# Create tip % col
df1['tip_percent'] = round(df1['tip_amount'] / (df1['total_amount'] - df1['tip_amount']),3)
```

Next, let's create another column called `generous`. This will be the target variable. The column should be a binary indicator of whether or not a customer tipped â‰¥ 20% (0=no, 1=yes).

```python
# Create 'generous' col (target)
df1['generous'] = df1['tip_percent']
df1['generous'] = (df1['generous'] >= 0.2)
df1['generous'] = df1['generous'].astype(int)
```

##### Create day column

Next, I'm going to be working with the pickup and dropoff columns.

For that purpose, I'll convert the `tpep_pickup_datetime` and `tpep_dropoff_datetime` columns to datetime.

```python
# Convert pickup and dropoff cols to datetime
df1['tpep_pickup_datetime'] = pd.to_datetime(df1['tpep_pickup_datetime'], format='mixed')
df1['tpep_dropoff_datetime'] = pd.to_datetime(df1['tpep_dropoff_datetime'], format='mixed')
```

Let's create a `day` column that contains only the day of the week when each passenger was picked up, converted to lowercase.

```python
# Create a 'day' col
df1['day'] = df1['tpep_pickup_datetime'].dt.day_name().str.lower()
```

#### Create time of day columns

Next, I should engineer four new columns that represent time of day bins. Each column should contain binary values (0=no, 1=yes) that indicate whether a trip began (picked up) during the following times:

`am_rush` = [06:00-10:00)  
`daytime` = [10:00-16:00)  
`pm_rush` = [16:00-20:00)  
`nighttime` = [20:00-06:00)  

```python
# Create 'am_rush' col
df1['am_rush'] = df1['tpep_pickup_datetime'].dt.hour

# Create 'daytime' col
df1['daytime'] = df1['tpep_pickup_datetime'].dt.hour

# Create 'pm_rush' col
df1['pm_rush'] = df1['tpep_pickup_datetime'].dt.hour

# Create 'nighttime' col
df1['nighttime'] = df1['tpep_pickup_datetime'].dt.hour
```

I'll need to write four functions to convert each new column to binary (0/1). I'll begin with `am_rush` - if the hour is between [06:00â€“10:00), it returns 1, otherwise, it returns 0.

```python
# Define 'am_rush()' conversion function [06:00â€“10:00)
def am_rush(hour):
    if 6 <= hour['am_rush'] < 10:
        val = 1
    else:
        val = 0
    return val
```

Next, I'll apply the `am_rush()` function to the `am_rush` series to perform the conversion, and print the first five values of the column to make sure it did what I expected it to do.

```python
# Apply 'am_rush' function to the 'am_rush' series
df1['am_rush'] = df1.apply(am_rush, axis=1)
df1['am_rush'].head()
```

Next, I should write three more functions to convert the remaining columns and apply them to their respective series.

```python
# Define 'daytime()' conversion function [10:00â€“16:00)
def daytime(hour):
    if 10 <= hour['daytime'] < 16:
        val = 1
    else:
        val = 0
    return val
```

```python
# Apply 'daytime' function to the 'daytime' series
df1['daytime'] = df1.apply(daytime, axis=1)
df1['daytime'].head()
```

```python
# Define 'pm_rush()' conversion function [16:00â€“20:00)
def pm_rush(hour):
    if 16 <= hour['pm_rush'] < 20:
        val = 1
    else:
        val = 0
    return val
```

```python
# Apply 'pm_rush' function to the 'pm_rush' series
df1['pm_rush'] = df1.apply(pm_rush, axis=1)
df1['pm_rush'].head()
```

```python
# Define 'nighttime()' conversion function [20:00â€“06:00)
def nighttime(hour):
    if 20 <= hour['nighttime'] < 24:
        val = 1
    elif 0 <= hour['nighttime'] < 6:
        val = 1
    else:
        val = 0
    return val
```

```python
# Apply 'nighttime' function to the 'nighttime' series
df1['nighttime'] = df1.apply(nighttime, axis=1)
df1['nighttime'].head()
```

##### Evaluation metric

Before modeling, I must decide on an evaluation metric. 
Therefore, I'll examine the class balance of the target variable. 

```python
# Get class balance of 'generous' col
df2['generous'].value_counts(normalize=True)
```

A little over half of the customers in this dataset were "generous" (tipped â‰¥ 20%). The dataset is very nearly balanced.

To determine a metric, I must consider the cost of both kinds of model error:
* False positives (the model predicts a tip â‰¥ 20%, but the customer does not give one)
* False negatives (the model predicts a tip < 20%, but the customer gives more)

False positives are worse for cab drivers, because they would pick up a customer expecting a good tip and then not receive one, frustrating the driver.

False negatives are worse for customers, because a cab driver would likely pick up a different customer who was predicted to tip more, even when the original customer would have tipped generously.

**The stakes are relatively even. I want to help taxi drivers make more money, but I don't want this to anger customers. Therefore, the metric should weigh both precision and recall equally. This metric is F1 Score that places equal weight on true postives and false positives, and so therefore on precision and recall.

### **Task 3. Modeling**

##### **Splitting the data**

Now I'm ready to model. The only remaining step is to split the data into features/target variable and training/testing data. 


```python
# Isolate target variable (y)
y = df2['generous']

# Isolate the features (X)
X = df2.drop('generous', axis=1)

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
```

##### **Random forest**

Let's begin with using `GridSearchCV` to tune a random forest model.

First, I'll instantiate the random forest classifier `rf` and set the random state.

Second, I'll create a dictionary `cv_params` of any of the following hyperparameters and their corresponding values to tune. The more I tune, the better my model will fit the data, but the longer it will take. 
 - `max_depth`  
 - `max_features`  
 - `max_samples` 
 - `min_samples_leaf`  
 - `min_samples_split`
 - `n_estimators`  

Third, I'll define a set `scoring` of scoring metrics for GridSearch to capture (precision, recall, F1 score, and accuracy).

Forth, I'll instantiate the `GridSearchCV` object `rf1`, and pass to it as arguments:
 - estimator=`rf`
 - param_grid=`cv_params`
 - scoring=`scoring`
 - cv=4 (I'll define 4 cross-validation folds)
 - refit=f1 (I'll choose the F1 Score as evaluation metric to use to select the model

```python
# 1. Instantiate the random forest classifier
rf = RandomForestClassifier(random_state=42)

# 2. Create a dictionary of hyperparameters to tune 
# First I will assign only one 1 value for each parameter for simplicity,
# and then I'll eventually experiment with ranges of values
cv_params = {'max_depth': [None],
             'max_features': [1.0],
             'max_samples': [0.7],
             'min_samples_leaf': [1],
             'min_samples_split': [2],
             'n_estimators': [300]
             }

# 3. Define a set of scoring metrics to capture
scoring = ['accuracy', 'precision', 'recall', 'f1']

# 4. Instantiate the GridSearchCV object
rf1 = GridSearchCV(rf, cv_params, scoring=scoring, cv=4, refit='f1')
```

```python
import pickle 
```

```python
def write_pickle(model_object, save_name:str):
    '''
    save_name is a string.
    '''
    with open(save_name + '.pickle', 'wb') as to_write:
        pickle.dump(model_object, to_write)
```

```python
def read_pickle(saved_model_name:str):
    '''
    saved_model_name is a string.
    '''
    with open(saved_model_name + '.pickle', 'rb') as to_read:
        model = pickle.load(to_read)

        return model
```

I'll write custom function `make_results()` to output all of the scores of the model. 

```python
def make_results(model_name:str, model_object, metric:str):
    '''
    Arguments:
    model_name (string): what I want the model to be called in the output table
    model_object: a fit GridSearchCV object
    metric (string): precision, recall, f1, or accuracy

    Returns a pandas df with the F1, recall, precision, and accuracy scores
    for the model with the best mean 'metric' score across all validation folds.
    '''

    # Create dictionary that maps input metric to actual metric name in GridSearchCV
    metric_dict = {'precision': 'mean_test_precision',
                 'recall': 'mean_test_recall',
                 'f1': 'mean_test_f1',
                 'accuracy': 'mean_test_accuracy',
                 }

    # Get all the results from the CV and put them in a df
    cv_results = pd.DataFrame(model_object.cv_results_)

    # Isolate the row of the df with the max(metric) score
    best_estimator_results = cv_results.iloc[cv_results[metric_dict[metric]].idxmax(), :]

    # Extract Accuracy, precision, recall, and f1 score from that row
    f1 = best_estimator_results.mean_test_f1
    recall = best_estimator_results.mean_test_recall
    precision = best_estimator_results.mean_test_precision
    accuracy = best_estimator_results.mean_test_accuracy

    # Create table of results
    table = pd.DataFrame({'model': [model_name],
                        'precision': [precision],
                        'recall': [recall],
                        'F1': [f1],
                        'accuracy': [accuracy],
                        },
                       )

    return table
```

Next, let's use the model to predict on the test data, and assign the results to a variable called `rf_preds`.

For this project milestone, I will use several models to predict on the test data. This decision always comes with a trade-off.

The benefit of using multiple models to predict on the test data is that I can compare models using data that was not used to train/tune hyperparameters. This reduces the risk of selecting a model based on how well it fit the training data.

The drawback of using the final test data to select a model is that, by using the unseen data to make a decision about which model to use, I no longer have a truly unbiased idea of how the model would be expected to perform on new data. In this case, I would think of final model selection as another way of "tuning" the model.

```python
# Get scores on test data
rf_preds = rf1.best_estimator_.predict(X_test)
```

In order to generate the scores on the test data, I'll write `get_test_scores()` function, and assign the results to `rf_test_scores`.


```python
def get_test_scores(model_name:str, preds, y_test_data):
    '''
    Generates a table of test scores.

    In:
    model_name (string): My choice: how the model will be named in the output table
    preds: numpy array of test predictions
    y_test_data: numpy array of y_test data

    Out:
    table: a pandas df of precision, recall, f1, and accuracy scores for the model
    '''
    accuracy = accuracy_score(y_test_data, preds)
    precision = precision_score(y_test_data, preds)
    recall = recall_score(y_test_data, preds)
    f1 = f1_score(y_test_data, preds)

    table = pd.DataFrame({'model': [model_name],
                        'precision': [precision],
                        'recall': [recall],
                        'F1': [f1],
                        'accuracy': [accuracy]
                        })

    return table
```
